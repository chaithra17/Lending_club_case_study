```python
## Outliers are present in most of the numerical columns
```

```python
for i in data.select_dtypes(include=['float64','int64']).columns:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 -(1.5*IQR)
    upper = Q3 + (1.5*IQR)
    data[i]=np.where(data[i]>upper,upper,np.where(data[i]<lower,lower,data[i]))
```
